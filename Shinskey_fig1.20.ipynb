```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
k_taod_tao1 = np.arange(0.1, 4, 0.1)
```

```python
print(k_taod_tao1)
```

```python
len(k_taod_tao1)
```

```python
List_G1 = np.zeros(len(k_taod_tao1))
```

```python
R_taon_taod = np.zeros(len(k_taod_tao1))
```

```python
List_phi1 = np.zeros(len(k_taod_tao1)); List_phid = np.zeros(len(k_taod_tao1)); pi=math.pi
```

```python
tol = 0.01; R = 4;
List_step = range(0, len(k_taod_tao1));
for step in List_step:
    loop=0;
    error = 1;
    if step>=0: R = R_taon_taod[step-1];
    k = k_taod_tao1[step];

    while(error>tol):
        phi1 = np.arctan(2*pi/(R*k));
        phid = pi - phi1;
        R_nxt = 2*pi/phid;
        error = abs(R-R_nxt);
        print("phi1=", phi1, "phid=", phid, "R=", R, "R_nxt=", R_nxt)
        R=R_nxt;
        loop = loop +1;

    print("step =", step, "max loop =", loop)
    List_phi1[step] = phi1; List_phid[step] = phid; List_G1[step] = np.cos(phi1); R_taon_taod[step]=R;
```

```python
plt.plot(k_taod_tao1, List_G1)
```

```python
plt.plot(k_taod_tao1, R_taon_taod)
```
